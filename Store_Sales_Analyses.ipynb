```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/Dataset/ShadowFox_Sales.csv")
df.head()
```

```python
df = df.dropna(subset=['Postal Code'])
```

```python
df.isnull().sum().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
```

```python
df.head()
```

```python
# monthly sales
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
```

```python
# create date column for plotting
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))

monthly_sales.head()
```

```python
import plotly.express as px

fig = px.line(monthly_sales, x='Date', y='Sales', title='Monthly Sales Trend')
fig.show()
```

```python
# category-wise sales
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

category_sales
```

```python
import plotly.express as px

fig = px.bar(category_sales, x='Category', y='Sales', title='Sales by Category')
fig.show()
```

```python
# sub-category wise sales
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().reset_index()

# sort descending
subcat_sales = subcat_sales.sort_values(by='Sales', ascending=False)

subcat_sales.head()
```

```python
fig = px.bar(subcat_sales, x='Sub-Category', y='Sales',
             title='Sales by Sub-Category')

fig.show()
```

```python
# profit column check
df.columns
```

```python
# segment-wise sales
segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()

segment_sales
```

```python
fig = px.pie(segment_sales, names='Segment', values='Sales',
             title='Sales by Customer Segment')
fig.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

fig = px.bar(region_sales, x='Region', y='Sales',
             title='Sales by Region')
fig.show()
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().reset_index()

top_customers = top_customers.sort_values(by='Sales', ascending=False).head(10)

top_customers
```

```python
# average sales per order
order_sales = df.groupby('Order ID')['Sales'].sum().reset_index()

order_sales['Avg Sales'] = order_sales['Sales']

order_sales.head()
```
